```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de la tabla 5.6
modelos = ['yolov8s', 'yolov9s', 'yolov10s', 'yolov11s', 'yolov12s', 'yolov12l', 'custom']
precision = [0.836, 0.852, 0.854, 0.845, 0.853, 0.841, 0.836]
precision_err = [0.008, 0.015, 0.014, 0.025, 0.011, 0.002, 0.014]
recall = [0.868, 0.866, 0.855, 0.869, 0.861, 0.867, 0.865]
recall_err = [0.015, 0.013, 0.011, 0.013, 0.016, 0.013, 0.014]
mAP50 = [0.914, 0.919, 0.918, 0.922, 0.918, 0.912, 0.905]
mAP50_err = [0.011, 0.009, 0.009, 0.012, 0.008, 0.011, 0.011]
mAP5095 = [0.723, 0.729, 0.739, 0.742, 0.730, 0.724, 0.707]
mAP5095_err = [0.024, 0.026, 0.029, 0.025, 0.018, 0.013, 0.015]

# Posiciones para grficos en y
y_pos = np.arange(len(modelos))

plt.figure(figsize=(12, 8))
plt.errorbar(precision, y_pos + 0.3, xerr=precision_err, fmt='o', label='Precisin', capsize=5)
plt.errorbar(recall, y_pos + 0.1, xerr=recall_err, fmt='o', label='Recall', capsize=5)
plt.errorbar(mAP50, y_pos - 0.1, xerr=mAP50_err, fmt='o', label='mAP50', capsize=5)
plt.errorbar(mAP5095, y_pos - 0.3, xerr=mAP5095_err, fmt='o', label='mAP50-95', capsize=5)

plt.yticks(y_pos, modelos)
plt.xlabel('Valor y rango de incertidumbre')
plt.title('Comparacin de mtricas y sus intervalos de incertidumbre para modelos YOLO')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.gca().invert_yaxis()  # Para que el modelo 'yolov8s' est arriba
plt.show()
```
